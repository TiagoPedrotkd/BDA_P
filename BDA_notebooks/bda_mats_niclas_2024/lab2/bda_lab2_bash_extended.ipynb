We can achieve similar results using ``mlr`` and "then"-chaining. Note that this is equivalent to calling ``| ./mlr *verb*`` sequentially.

```python
%%bash
 
./mlr --from sales.csv --csv filter '$Region == "Europe" && $["Sales Channel"] == "Online"' \
    then count-distinct -f Country -o Count \
    then sort -nr Count \
    then head -n 5
```

We can also generate summary statistics using ``mlr`` using stats1 for univariate data.

```python
%%bash

./mlr --from sales.csv --csv filter '$Region == "North America"' \
    then count-distinct -f Country -o Count \
    then stats1 -a min,p10,p50,p90,max -f Count \
    then head -n 5
```

We can put all of it together in a single command, using the ";" delimiter. We first find all the "html" files in the current directory and re-direct them into another file. Then, we perform parallel counts of the number of rows contained in any of those files. We use 5 jobs at the same time. Finally, we print the name of the file and use "pup" to count the number of rows in the table.

```python
# Saving a new file with the HTML files
!find . -name "*.html" > html_sites.txt; parallel -j 5 'echo {} && cat {} | ./pup -n "table.wikitable tr"' :::: html_sites.txt
```

## Extra: Parallel Processing and mlr

``mlr`` can be relatively slow to perform queries with. We can use parallel processing to speed them up. We can do this using the `parallel` command and by splitting the file to be queried against. 

```python
# This is the base query using plain mlr
! ./mlr --csv filter '$Country=="Portugal"' sales.csv > portugal_filtered.csv
```

```python
%%bash

# Extract header from file
head -n 1 sales.csv > header.csv

# Split the content of the file. The l/ ensures that lines are not broken. The -d option creates numeric suffixes. -a controls the length of the numeric suffix. Using "sales_" after the input file, we specify the split prefix.
split --filter='tail -n +2 > $FILE' sales.csv sales_ -nl/10 -a 1 -d

# Combine header with each split part. Using the bracketed parentheses, we can separate the interpolated $file from a custom extension
for file in sales_*; do cat header.csv "$file" > "${file}_with_header.csv"; done

# Remove temporary header file
rm header.csv
```

```python
%%bash

echo """
#!/bin/bash

filename=\$1 # This is a variable assignment
country_name=\$2

# Since we are in a script we have to escape quotes
./mlr --csv filter '\$Country==\"\${country_name}\"' \"\$filename\" > \"\${filename%.*}_\${country_name}_filtered.csv\"
""" > script.sh
```

```python
# This command makes script.sh executable
!chmod +x script.sh
```

Since the free version of DataBricks only has 2 cores, this will not perform very will. As more cores are added to the machine, a significant speed-up should be achieved. On a single core outside of DataBricks, the initial ``mlr`` filter command takes around 10 seconds. By splitting the data first and then running filters on each file, the operation takes around 3 seconds. Note that there is an additional overhead by needing the to split the files. This comes down to another 3 seconds, in this example. Since the splitting operation is relatively cheap, in computational terms, further differences in performance should be noticed with larger file sizes and more available cores.

```python
!parallel -j 2 ./script.sh {} 'Portugal' ::: *with_header.csv
```
